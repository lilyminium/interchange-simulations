```python
import json
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
filename = "boxes-nosort_n-2000_interchange-equilibration"
```

```python
df = pd.read_csv(f"{filename}.csv")
df["t0 (ps)"] = df["t0"] * 2
```

```python
df
```

```python
g = sns.catplot(
    data=df,
    kind="box",
    y="property",
    x="t0 (ps)",
    aspect=2
)
g.savefig(f"images/{filename}.png", dpi=300)
```

```python
T0_PS = "t0 (ps)"
U = "Potential Energy (kJ/mole)"

df[
    (df[T0_PS] > 10000)
    & (df["property"] == U)
].sort_values(T0_PS, ascending=False)#.to_csv("long_equilibration_times.csv")
```

```python
with open("boxes-nosort/n-2000/liquid-boxes.json", "r") as f:
    boxes = json.load(f)
```

```python
sfe_ix = [
    i
    for i, x in enumerate(boxes)
    if 1 in x["n_molecules"]
]
```

```python
g = sns.catplot(
    data=df[df.entry.isin(sfe_ix)],
    kind="box",
    y="property",
    x="t0 (ps)",
    aspect=2
)
g.savefig(f"images/{filename}_sfes.png", dpi=300)
```

```python
df[
    (df[T0_PS] > 5000)
    & (df["property"] == U)
    & (df.entry.isin(sfe_ix))
].sort_values(T0_PS, ascending=False).to_csv("long_equilibration_times_sfe.csv")
```
